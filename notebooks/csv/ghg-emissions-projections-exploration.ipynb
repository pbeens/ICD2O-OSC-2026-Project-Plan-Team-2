# GHG Emissions Projections Exploration

**Objective:** Explore Canada's historical greenhouse gas emissions and projection scenarios with beginner-friendly steps.

**Dataset description:** This file includes yearly values from 2005 to 2040. It has one historical emissions column and three projection scenario columns. Each row represents one year.

**What are projection scenarios?**
Scenarios are different possible futures based on which policies and measures are included.
- Historical emissions: measured values from past years
- With measures: projection using measures already in place
- With additional measures: projection including extra planned measures

**Data source(s):**
- Dataset page: https://open.canada.ca/data/en/dataset/2ff017d8-e122-4d6b-a394-f919b64018cc
- English CSV: https://www.canada.ca/content/dam/eccc/documents/csv/cesindicators/ghg-projections/2025/ghg-emissions-projections-en.csv
- French CSV: https://www.canada.ca/content/dam/eccc/documents/csv/cesindicators/ghg-projections/2025/projections-emissions-ges-fr.csv

## Step 1: Import our libraries

- `pandas` (`pd`) helps us work with table data.
- `plotly.express` (`px`) helps us make charts.


```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the CSV and preview rows

A DataFrame is a table object in pandas. We call it `df` (short for data frame).
The file uses Latin-1 text encoding, so we include `encoding="latin1"` when loading.


```python
data_url = "https://www.canada.ca/content/dam/eccc/documents/csv/cesindicators/ghg-projections/2025/ghg-emissions-projections-en.csv"
df = pd.read_csv(data_url, encoding="latin1")
df.head()
```

## Step 3: Display column names

Before cleaning, check the current column names.


```python
display(df.columns)
```

## Step 4: Use row index 1 as the column names

This dataset stores the real column headers in row index `1`, not in the first header line pandas used.

How `df.iloc[1]` works:
- `iloc` means *integer location* (row/column position by number).
- `df.iloc[1]` means "give me the second row" (index starts at 0).
- In this file, that second row contains the real header text (`Year`, scenario names, etc.).

When we run `df.columns = df.iloc[1]`, pandas replaces the current column names with the values from that row.


```python
df.columns = df.iloc[1]
display(df.columns)
```

## Step 5: View the full table

Display the full DataFrame so we can see header rows, data rows, and note rows.


```python
display(df)
```

## Step 6: Check data types for all columns

This tells us how pandas currently reads each column (for example `object`, `int64`, or `float64`).
We check this before conversion so we can verify whether numeric conversion is needed for plotting.


```python
display(df.dtypes)
```

## Step 7: Convert all columns to numeric

This is a `for` loop. A loop repeats one action for each item in a list.
Here, `df.columns` is the list of column names. On each pass, `col` becomes one column name.
`errors="coerce"` turns non-numeric values into `NaN` instead of stopping with an error.


```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

display(df.dtypes)
```

## Step 8: Keep rows with valid years

Some rows in this file are notes/sources, not yearly data.
After numeric conversion, those note rows have `NaN` in `Year`, so we keep only rows where `Year` is present.


```python
df = df.dropna(subset=["Year"]).copy()
display(df.head())
display(df.tail())
```

## Step 9: Build a line chart (Historical + Projection Scenarios)


```python
fig = px.line(
    df,
    x="Year",
    y=[
        "Historical Emissions",
        "With Measures",
        "With Additional Measures (No NBCS/Agriculture)",
        "With Additional Measures"
    ],
    title="Canada GHG Emissions: Historical and Projection Scenarios"
)
fig.show()
```
