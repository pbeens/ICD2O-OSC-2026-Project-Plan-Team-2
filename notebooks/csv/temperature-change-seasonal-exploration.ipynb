# Temperature Change (Seasonal) Exploration

**Objective:** Explore seasonal temperature departure trends in Canada with beginner-friendly steps.

**Dataset description:** This file reports yearly temperature departures from a historical reference period (1961-1990). Each row is a year, and each seasonal column shows how much warmer or cooler that season was versus the reference.

**What is a seasonal temperature departure?**
A seasonal temperature departure is the difference between a season's observed average temperature and the baseline average temperature for that same season.
- Positive value = warmer than baseline
- Negative value = cooler than baseline

**Where does the baseline come from?**
The baseline is based on Environment and Climate Change Canada's 1961-1990 climate normal period. Seasonal averages from that period are used as the reference for comparison.

**Data source(s):**
- English CSV: https://www.canada.ca/content/dam/eccc/documents/csv/cesindicators/temperature-change/2019/Temperature-change-seasonal-en.csv
- French CSV: https://www.canada.ca/content/dam/eccc/documents/csv/cesindicators/temperature-change/2019/Temperature-change-seasonal-fr.csv

## Step 1: Import our libraries

- `pandas` (`pd`) helps us work with table data.
- `plotly.express` (`px`) helps us make charts.


```python
import pandas as pd
import plotly.express as px
```

## Step 2: Load the CSV and preview rows

A DataFrame is a table object in pandas. We call it `df` (short for data frame).
In `df.head()`, `head()` is a method (an action).


```python
data_url = "https://www.canada.ca/content/dam/eccc/documents/csv/cesindicators/temperature-change/2019/Temperature-change-seasonal-en.csv"
df = pd.read_csv(data_url, encoding="latin1")
df.head()
```

## Step 3: Display column names

Before cleaning, check the current column names.


```python
display(df.columns)
```

## Step 4: Rename columns

The original names include one long title and `Unnamed` placeholders.
We rename them to short names we can use easily.


```python
df.columns = ["Year", "Winter", "Spring", "Summer", "Autumn"]
display(df.columns)
```

## Step 5: View the full table

Display the full DataFrame so we can see what needs cleanup.


```python
display(df)
```

## Step 6: Delete rows with NaN values

`NaN` means Not a Number (missing/invalid data).
`dropna()` removes rows that contain `NaN`.
`copy()` makes a separate DataFrame object for safer follow-up edits.


```python
df = df.dropna().copy()
display(df)
```

## Step 7: Delete row index 1

Row index 1 still contains text labels instead of data values, so remove it.


```python
df = df.drop(index=1, errors="ignore").copy()
display(df)
```

## Step 8: Check data types for all columns
This tells us how pandas currently reads each column (for example `object`, `int64`, or `float64`).
We check this before conversion so we can verify whether numeric conversion is needed for plotting.


```python
display(df.dtypes)
```

## Step 9: Convert all columns to numeric
This is our first `for` loop. A loop repeats the same action for each item in a list.
Here, `df.columns` is the list of column names. On each pass, `col` becomes one column name (like `Year`, then `Winter`, and so on).
Inside the loop, we convert that column to numeric. `errors="coerce"` turns non-numeric values into `NaN` instead of stopping with an error.


```python
for col in df.columns:
    df[col] = pd.to_numeric(df[col], errors="coerce")

display(df.dtypes)
```

## Step 10: Build a line chart (All seasons vs Year)


```python
fig = px.line(
    df,
    x="Year",
    y=["Winter", "Spring", "Summer", "Autumn"],
    title="Canada Seasonal Temperature Departure by Year"
)
fig.show()
```
