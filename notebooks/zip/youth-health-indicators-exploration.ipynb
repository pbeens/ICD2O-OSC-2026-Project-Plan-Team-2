```python
import pandas as pd
import plotly.express as px
from io import BytesIO
from urllib.request import urlopen
from zipfile import ZipFile
```

```python
english_csv_url = "https://www150.statcan.gc.ca/n1/tbl/csv/13100947-eng.zip"

with urlopen(english_csv_url) as response:
    zip_bytes = response.read()

with ZipFile(BytesIO(zip_bytes)) as zf:
    csv_files = [name for name in zf.namelist() if name.lower().endswith('.csv')]
    print("CSV files in ZIP:", csv_files)
    data_file = next(name for name in csv_files if "metadata" not in name.lower())
    with zf.open(data_file) as f:
        df = pd.read_csv(f)

print(f"Loaded file: {data_file}")
print(f"Rows: {len(df)}")
print(f"Columns: {len(df.columns)}")
display(df.head())
```

```python
if df is not None:
    summary = pd.DataFrame({
        "column": df.columns,
        "dtype": [str(df[col].dtype) for col in df.columns],
        "null_count": [int(df[col].isna().sum()) for col in df.columns]
    })
    display(summary)
else:
    print("Dataset not loaded yet.")
```
