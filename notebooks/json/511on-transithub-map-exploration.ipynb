# 511ON Transit Hubs - JSON Map Exploration

**Objective:** Load Ontario transit hub records from a JSON data link and map each location using latitude and longitude.

**Dataset description:** Each row is one transit hub. In this lesson, we focus on the hub name, region, address, and map coordinates.

**Data source(s):**
- API endpoint: https://511on.ca/api/v2/get/transithub
- API documentation: https://511on.ca/help/endpoint/transithub

**Date/author:** 2026-03-15 / Codex


### Step 1: Import libraries
Big picture: load the tools we need for tables and charts.
Technical details: `pandas` handles DataFrames, and `plotly.express` creates the map chart.


```python
import pandas as pd
import plotly.express as px
```

### Step 2: Load the JSON data into a DataFrame
Big picture: download the transit hub data and preview the first few rows.
Technical details: `pd.read_json(...)` reads JSON from the link and builds a table named `df`. A DataFrame is a table object in pandas, and `df` is a short name for "data frame."


```python
url = "https://511on.ca/api/v2/get/transithub"
df = pd.read_json(url)
display(df.head())
```

### Step 3: Display the column names
Big picture: check what fields are available before we clean the table.
Technical details: `df.columns` is a property, which means it gives information about the DataFrame instead of performing an action.


```python
display(df.columns)
```

### Step 4: Keep only the columns we need
Big picture: focus on the most useful fields for mapping and simple analysis.
Technical details: we select a smaller set of columns and use `copy()` to make a separate DataFrame.


```python
hub_df = df[["Name", "Region", "Address", "Latitude", "Longitude", "Website"]].copy()
display(hub_df.head())
```

### Step 5: Remove rows with missing coordinates
Big picture: keep only rows that can be placed on a map.
Technical details: `dropna(...)` removes rows with missing values. `NaN` means a value is missing. In `subset=["Latitude", "Longitude"]`, pandas checks only those two columns.


```python
hub_df = hub_df.dropna(subset=["Latitude", "Longitude"]).copy()
display(hub_df.head())
```

### Step 6: Count the number of transit hubs
Big picture: check how many mapped transit hubs are in the dataset.
Technical details: `len(hub_df)` returns the number of rows in the table.


```python
len(hub_df)
```

### Step 7: Count hubs by region
Big picture: see which regions have the most transit hubs.
Technical details: `value_counts()` counts how many times each region appears, and `reset_index()` turns the result back into a table.


```python
region_counts = hub_df["Region"].value_counts().reset_index()
region_counts.columns = ["Region", "HubCount"]
display(region_counts)
```

### Step 8: Make a bar chart of hubs by region
Big picture: visualize which regions have more transit hubs.
Technical details: each bar shows the number of hubs in one region.


```python
chart_df = region_counts.sort_values("HubCount")

fig = px.bar(
    chart_df,
    x="HubCount",
    y="Region",
    orientation="h",
    title="Number of transit hubs by region",
    labels={"HubCount": "Hub count", "Region": "Region"},
)
fig.show()
```

### Step 9: Plot the transit hubs on a map
Big picture: place each transit hub on an interactive map of Ontario.
Technical details: `px.scatter_mapbox(...)` uses latitude and longitude from each row. Hover text shows the hub name and region.


```python
fig = px.scatter_mapbox(
    hub_df,
    lat="Latitude",
    lon="Longitude",
    hover_name="Name",
    hover_data={"Region": True, "Address": True, "Latitude": False, "Longitude": False, "Website": True},
    zoom=4.8,
    height=600,
    title="511ON transit hubs in Ontario",
)
fig.update_layout(mapbox_style="open-street-map")
fig.update_layout(margin={"r": 0, "t": 50, "l": 0, "b": 0})
fig.show()
```

Big picture: think about what this dataset can help us learn.
